### Save LowerData with the addition of Amd Data information from another article.

```python
lower_Er_U = CDP.lower
lower_Er_U.head(3)
```

```python
def term_concat(df):
    return df.TERM1+'-'+df.TERM2
```

```python
## Er A_MD (Probabilities for Radiative and Nonradiative Decay of Er'+ in LaF, Weber 1967)
df_Er_md = pd.read_csv('./Carnal_data/Nd_AMD_pre_n_2.16.txt',delimiter='\t')
df_Er_md['Term'] = df_Er_md.apply(lambda x:term_concat(x),axis=1)
df_Er_md['A_md_pre'] = df_Er_md.AMD_pre.apply(lambda x:x/(2.16**3))
df_Er_md.head(3)
```

```python
lower_Er_U = lower_Er_U.merge(right = df_Er_md[['Term','A_md_pre']], how='left', on='Term')
lower_Er_U = lower_Er_U.fillna(0)
```

```python
print(lower_Er_U.shape)
lower_Er_U.query('A_md_pre>0')
```

```python
lower_Er_U.to_csv(f'./U_data/Nd_Carnal_Lower.txt')
```
