## Сведенья об объекте
* Масса стекла ~50 грамм
* молярная доля как Nd2O3 в стекле 0.5 моль%
* Плотности стекол с Nd - 3,14;
* Показатель преломления на длинне волны 589  - 1.55
* N_A = 6.022*1e23
* Молярная масса ZnO - 81.389
* Молярная масса P2O5 - 141.943
* Молярная масса Nd2O3 - 336.477
* Моль% 57.00 ZnO 42.50 P2O5 0.5 Nd2O3

**Относительная молярная масса стекла с Nd2O3 (0.57*81.389 + 0.425*141.943 + 0.005 * 336.477)**
_____________

```python
## Определение концентрации Er (ions/cm3)
Md_Nd = 336.477*0.005/((0.57)*81.389 + 0.425*141.943 + 0.005 * 336.477)
print(f'Массовая доля Nd2O3  {Md_Nd}')
Ny_Nd = 50*Md_Nd/336.477*2
print(f'Кол-во вещества Nd в стекле Nd2O3 {Ny_Nd}')
N_A = 6.02214076e23
c_ND = Ny_Nd * N_A *3.14 / 50
print(f'Nd concantration = {c_ND} (ions/cm3)')
```
